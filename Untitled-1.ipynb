```python
# Loading of libraries and documents
from nltk.stem import PorterStemmer, WordNetLemmatizer
import nltk
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
import string
from nltk.corpus import stopwords
import math
from collections import Counter
from operator import itemgetter
nltk.download('stopwords')
nltk.download('punkt')
```

```python
stemmer = PorterStemmer() 

def tokenize(text):
    """
    It tokenizes and stems an input text.
    
    :param text: str, with the input text
    :return: list, of the tokenized and stemmed tokens.
    """
    text = "".join([ch for ch in text if ch not in string.punctuation])
    tokens = nltk.word_tokenize(text)

    # remove punctuation

    return [stemmer.stem(word.lower()) for word in tokens if word not in stopwords.words('english')]
```

```python
# Read a list of documents from a file. Each line in a file is a document
with open("a/20191031.txt") as f:
    fcontent = f.read()
    
    content = fcontent.split("Form 10-K",1)
    content = content[1:]

document = [tokenize(d) for d in content]

document = document[0]
```

```python
# Read a list of documents from a file. Each line in a file is a document
with open("words base.txt") as f:
    fcontent = f.read()
    
    vcontent = fcontent.split(", ")

vocab = [x.strip() for x in vcontent] 
```

```python
def bag_of_words(text, given_words):
    word_count = Counter(text)
    bow = {word: word_count[word] for word in given_words}
    return bow
```

```python
bow = bag_of_words(document, vocab)
```

```python
# Reference code using scikit-learn
tf = TfidfVectorizer(tokenizer = tokenize,analyzer='word', ngram_range=(1,1), min_df = 1)
features = tf.fit_transform(content)
npm_tfidf = features.todense()
new_features = tf.transform(vocab)

cosine_similarities = linear_kernel(new_features, features).flatten()
```
